```python
"""The module contain the function to enable a user export the pdb file to a different folder with a new name."""

def Store_file(filepath):
    """This function exports and stores the current loaded pdb file into a new file in the defined path.
    Arguments: User inputs path and the new pdb filename"""
    
    import os
    out_path = input ("""
     Enter an authentic path to your file destination and the new filename.pdb 
     :""")

    # Split the path to obtain the file directory and new filename
    out_path1 = out_path.split("/")
    directory = "/".join(out_path1[:-1])
    filename = out_path1[-1]

    name_tag = False
    # Checking if path exists
    if os.path.isdir(directory):
 
        extension = filename.split(".")

        if len(extension) == 1: # checking if the extention is included
            filename = filename + ".pdb"
            out_path = directory + "/" + filename
            name_tag = True

        else:
            if len(extension) == 2 and extension[-1] == 'pdb': # confirm right extension name
                name_tag = True
            else:
                print("Check your output filename extension and name it appropriately")
                Store_file(filepath)
                
        if os.path.exists(out_path):
            print('The file %s exists!' %out_path)
            Store_file(filepath)

        if name_tag:
            #open file for writing and export
            with open(out_path, "w") as myfile:

                pdbfile = open(filepath,"r")
                for line in pdbfile:
                    myfile.write(line) # write to new file

                pdbfile.close() 

                print("Successfully exported the pdb file to %s" %out_path)
        
    else:
        print("The directory does not exist ")
```

```python
def Read_file(infile):
    
    """ This function loads a pdb file and reads contents for analysis """
    
    import os
    global path
    
    filepath = input('Enter an authentic path to a pdb file: ')
    
    # Split the path to obtain the file directory and file name
    filepath1 = filepath.split('/')
    directory = '/'.join(filepath1[:-1])
    filename = filepath1[-1]
    
    # Checking if filepath and file are true
    if os.path.isdir(directory):
        if os.path.isfile(filepath):
            myfile = open(filepath,'r')
            line_1 = myfile.readline()
            lastline = myfile.readlines()[-1]
            myfile.close()
            #all pdb files start with header and have a length of 81 and lastline starts with end
            if line_1.startswith('HEADER') and len(line_1) == 81 and lastline.startswith('END'): 
                print('The File %s has been successfully loaded' %filename)
                Store_file(filepath)
#                 pdb = filename # reassign the name of the pdb file to update on menu
            else:
                print('This is not a pdb file')
                
        else:
            extension = filename.split('.')
            if extension[-1] == 'pdb':
                print('The File %s does not exist' %filename)
                
            else:
                print('%s is not a pdb file' %filename)
                
    else:
        print('The directory does not exist!')
     
    return infile, filepath
```
